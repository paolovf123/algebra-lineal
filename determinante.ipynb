```python

#generamos nuestra matriz, los calculos de los autovectores
A = np.array([[0,0,1,0],[0,1,0,0],[0,1,1,0],[0,0,0,1]])
# la funci√≥n de abajo nos da los valores propios, autovectores
lambdas, V = np.linalg.eig(A)
print(V)
print(lambdas)
print(A[lambdas==0,:])
B = np.array([[0,1,0,1],[1,0,2,0],[0,0,1,0],[0,1,1,5]])

#A no tiene inversa por que sus autovectores no son linealmente independientes
```

```python
#para la determinante y la inversa tenemos
print(np.linalg.inv(B.T))
print(np.linalg.det(B.T))
```

```python
##para la norma
print(np.linalg.norm(B[0]))
```

```python
##tipos de normas
vector = np.array([-1,-2,0,4,5,6])
vector1 = np.array([1,3,2,4,5,6])
#normla L0 nos da los valores distintos de 0
print(np.linalg.norm(vector, ord=0))
#norma L1 nos da la suma de los valores absolutos
print(np.linalg.norm(vector, ord=1))
#norma L2 nos da la distancia al origen
print(np.linalg.norm(vector, ord=2))
#norma L2 al cuadrado es un vector por us transpuesta(producto interno)
print(pow((np.linalg.norm(vector, ord=2)),2))
print(np.dot(vector,vector.T))
#norma L_infinito nos da el valor maximo de nuestro arreglo
print(np.linalg.norm(vector, ord=np.inf))
```

```python
#producto interno
print(np.dot(vector,vector1.T))
print(np.linalg.norm(vector))
print(np.linalg.norm(vector1))
```
