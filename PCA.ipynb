```python
from sklearn.decomposition import PCA
```

```python
data_pca = PCA(n_components = 0.999)
data_pca.fit(data.data)
```

```python
filas = 3
columnas = data_pca.n_components_ // filas
fig, axes = plt.subplots(filas, columnas , figsize=(12,6),
                         subplot_kw = { 'xticks' : [], 'yticks' : []},
                         gridspec_kw = dict(hspace =0.01, wspace = 0.01))

for i, ax in enumerate(axes.flat):
    ax.imshow(data_pca.components_[i].reshape(64,64), cmap = "grey"),
```

```python
print(data_pca.n_components_)
```

```python
componentes = data_pca.transform(data.data)
proyeccion = data_pca.inverse_transform(componentes)
```

```python
fig, axes = plt.subplots(5, 10 , figsize=(12,6),
                         subplot_kw = { 'xticks' : [], 'yticks' : []},
                         gridspec_kw = dict(hspace =0.01, wspace = 0.01))

for i, ax in enumerate(axes.flat):
    ax.imshow(proyeccion[i].reshape(64,64), cmap = "grey"),
```
